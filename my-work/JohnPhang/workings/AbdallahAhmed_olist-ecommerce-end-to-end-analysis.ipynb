# Chapter 1: The Great Merger & Spotting the Gaps 
**"We will use a Left Join to keep all orders, even those with missing details, to understand the integrity of our dataset before we start cleaning."**

```python
orders = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_orders_dataset.csv')
items = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_items_dataset.csv')
customers = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_customers_dataset.csv')
products = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_products_dataset.csv')
payments = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_payments_dataset.csv')
reviews = pd.read_csv('/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_reviews_dataset.csv')
```

```python
df = orders.merge(items, on='order_id', how='left') \
           .merge(customers, on='customer_id', how='left') \
           .merge(products, on='product_id', how='left') \
           .merge(payments, on='order_id', how='left') \
           .merge(reviews, on='order_id', how='left')

df
```

# **Chapter 2: The Cleaning Operation**
**"Data cleaning is the most critical step in the pipeline. We found that our dates are stored as text, and several columns have missing values. In this chapter, we will fix these issues to ensure our future insights are accurate."**

```python
date_columns = [
    'order_purchase_timestamp', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_delivered_customer_date', 
    'order_estimated_delivery_date'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col])

df[date_columns].dtypes
```

**"Not all missing values are errors. For example, a missing delivery date might mean the order is still in transit or was canceled. We will fill missing review text with 'No Review' and remove only the rows that lack essential price information."**

```python
# 1. Fill missing review comments and titles
df['review_comment_title'] = df['review_comment_title'].fillna('No Title')
df['review_comment_message'] = df['review_comment_message'].fillna('No Message')
```

```python
# 2. Fill missing product categories with 'other'
df['product_category_name'] = df['product_category_name'].fillna('other')
```

```python
# 3. Drop rows where we have NO product_id or price (833 rows)
# Because we cannot analyze a sale without a price!
df.dropna(subset=['product_id', 'price'], inplace=True)
```

```python
print("Missing values handled!")
print(f"Current rows in dataset: {len(df)}")
```

**"Even after initial cleaning, some missing values remain. However, these are Logical Gaps. For example, missing delivery dates (order_delivered_customer_date) occur because some orders are still 'shipped' or 'invoiced' but not yet 'delivered'. Similarly, missing reviews mean the customer chose not to leave feedback. We will treat these carefully to maintain data integrity."**

```python
df.isnull().sum()
```

```python
df.dropna(subset=['payment_type'], inplace=True)
```

```python
cols_to_fix = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for col in cols_to_fix:
    df[col] = df[col].fillna(df[col].median())
```

```python
df['product_name_lenght'] = df['product_name_lenght'].fillna(0)
df['product_description_lenght'] = df['product_description_lenght'].fillna(0)
df['product_photos_qty'] = df['product_photos_qty'].fillna(0)
```

```python
print("Remaining logical gaps handled!")
print(f"Final Row Count for Chapter 2: {len(df)}")
```

# Chapter 3: Outlier Detection & Business Anomalies
**"Now that our data is clean and structured, we need to ensure its statistical reliability. In this chapter, we will use the Interquartile Range (IQR) method to detect price outliers. We will also investigate 'Business Anomalies', such as orders where the shipping cost (Freight) exceeds the product price."**

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

price_outliers = df[df['price'] > upper_bound]

print(f"Statistical Summary for Price:")
print(f"Upper Bound (Max logical price): ${upper_bound:.2f}")
print(f"Number of Outliers: {len(price_outliers)} out of {len(df)}")
print(f"Percentage of Outliers: {(len(price_outliers)/len(df))*100:.2f}%")
```

**"Beyond statistics, we must look at Business Logic. We will flag orders where the freight value is higher than the product price, as these cases might indicate logistics issues or data entry errors."**

```python
high_freight_orders = df[df['freight_value'] > df['price']]

print(f"--- Business Anomaly Report ---")
print(f"Orders with Shipping > Price: {len(high_freight_orders)}")
print(f"Max Freight Value found: ${df['freight_value'].max():.2f}")
```

# Chapter 6: RFM Customer Segmentation
**"Not all customers are equal. RFM helps us identify our Champions
and our At-Risk customers so we can act accordingly."**

**What is RFM?**

RFM stands for Recency, Frequency, and Monetary — three behavioral signals that together tell us how valuable each customer is.

• Recency: how many days since their last purchase? (lower = better)
• Frequency: how many unique orders did they place?
• Monetary: how much did they spend in total?

We use customer_unique_id instead of customer_id because Olist assigns a new customer_id to the same person every time they order — so customer_id would make every customer look like a one-time buyer.

```python
reference_date = df['order_purchase_timestamp'].max() + timedelta(days=1)

rfm = df.groupby('customer_unique_id').agg(
    Recency   = ('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    Frequency = ('order_id', 'nunique'),
    Monetary  = ('price', 'sum')
).reset_index()
```

**Why quartiles?**

We divide each metric into 4 equal groups (quartiles) and assign a score from 1 to 4. This makes customers comparable regardless of the actual numbers.

Note that Recency is reversed — a customer who bought 2 days ago is better than one who bought 200 days ago, so they get score 4, not 1.

We use rank(method='first') on Frequency and Monetary to handle duplicate values cleanly without errors.

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'],   q=4, labels=[4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first'),  q=4, labels=[1, 2, 3, 4])

rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

From scores to segments.
We combine the three scores to classify each customer into a business-meaningful group. This is where data becomes strategy.

```python
def assign_segment(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])
    if r == 4 and f == 4:
        return 'Champion'
    elif r >= 3 and f >= 3:
        return 'Loyal Customer'
    elif r == 4:
        return 'Recent Customer'
    elif r == 2 and f >= 2:
        return 'At Risk'
    elif r <= 1:
        return 'Lost'
    else:
        return 'Potential Loyalist'

rfm['Segment'] = rfm.apply(assign_segment, axis=1)
```

```python
seg_counts = rfm['Segment'].value_counts().reset_index()
seg_counts.columns = ['Segment', 'Count']

fig_rfm = px.bar(seg_counts, 
                 x='Segment', y='Count',
                 color='Segment',
                 title='Customer Segmentation (RFM Analysis)',
                 text='Count',
                 color_discrete_sequence=px.colors.qualitative.Bold)

fig_rfm.update_traces(textposition='outside')
fig_rfm.update_layout(showlegend=False)
fig_rfm.show()
```

The RFM segmentation reveals a critical challenge: Lost customers (23,748) represent the largest segment — nearly 4x the size of our Champions (6,075). This means for every loyal, high-value customer Olist has retained, almost four have already churned.
The middle segments (At Risk: 17,929 / Loyal: 17,906 / Potential Loyalists: 17,841) are remarkably balanced in size, suggesting a natural customer lifecycle funnel where customers gradually disengage over time rather than dropping off suddenly.
Actionable Strategies:

Champions (6,075): Small but gold. Invest in loyalty rewards and early-access offers to keep them engaged — they're your highest LTV customers.
Potential Loyalists (17,841): The biggest opportunity. One well-timed personalized campaign could convert them into the Loyal or Champion tier.
At Risk (17,929): Launch a re-engagement campaign with a targeted discount before they join the Lost segment.
Lost (23,748): Don't over-invest here. Acquisition cost of new customers is likely lower than win-back cost at this scale.

# Chapter 7: Monthly Revenue Trend

"To understand business health over time, we need to move from individual transactions to monthly aggregates. We convert timestamps to monthly periods and calculate total revenue and unique orders per month."

```python
df['year_month'] = df['order_purchase_timestamp'].dt.to_period('M')

monthly_revenue = df.groupby('year_month').agg(
    revenue = ('price', 'sum'),
    orders  = ('order_id', 'nunique')
).reset_index()

monthly_revenue['year_month'] = monthly_revenue['year_month'].astype(str)
```

"A single month's revenue can be misleading due to seasonal spikes. The 3-month rolling average smooths out this noise and reveals the true underlying trend."

```python
monthly_revenue['rolling_avg'] = (
    monthly_revenue['revenue']
    .rolling(window=3, min_periods=1)
    .mean()
)
```

```python
fig_revenue = go.Figure()

fig_revenue.add_trace(go.Bar(
    x=monthly_revenue['year_month'],
    y=monthly_revenue['revenue'],
    name='Monthly Revenue',
    marker_color='#5DCAA5',
    opacity=0.7
))

fig_revenue.add_trace(go.Scatter(
    x=monthly_revenue['year_month'],
    y=monthly_revenue['rolling_avg'],
    name='3-Month Rolling Avg',
    line=dict(color='#D85A30', width=2.5),
    mode='lines'
))

fig_revenue.update_layout(
    title='Monthly Revenue Trend with 3-Month Rolling Average',
    xaxis_title='Month',
    yaxis_title='Revenue (R$)',
    hovermode='x unified',
    legend=dict(orientation='h', yanchor='bottom', y=1.02)
)

fig_revenue.show()
```

The revenue chart tells a clear two-act story.
Act 1 — Explosive Growth (Jan 2017 → Jan 2018): Revenue grew consistently from near zero to over R$1M/month, representing roughly 10x growth in just 12 months. The 3-month rolling average confirms this wasn't seasonal noise — it was a genuine upward trend.
Act 2 — The Drop (Aug 2018 onwards): Revenue falls sharply in the final months. However, this is almost certainly a data artifact, not a real decline — the dataset was collected up to August 2018, so the last 1-2 months have incomplete orders that haven't been fully recorded yet. This should be noted explicitly to avoid misleading conclusions.
Key Observations:

Peak month: November 2017 — likely driven by Black Friday, which is massive in Brazil.
Plateau phase (Mar–Jul 2018): Revenue stabilized around R$900K–R$1M, suggesting market maturation in that period.
Recommendation: Olist should investigate whether the plateau indicates a growth ceiling or a temporary saturation, and double down on the Black Friday spike with early campaign planning.
